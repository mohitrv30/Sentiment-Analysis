# VADER sentiments for family

# obtaining list of sentence

```python

def find_sentence(review):
    coverage=['coverage','plan','care','policy','corporate']
    temp = []
    for word in coverage:
        blob = TextBlob(review)
        sentences = blob.sentences
        for sentence in sentences:
            if word in sentence.split():
                temp.append(sentence)
    temp1 = ''
    for i in temp:
        temp1 = temp1+str(i)
    return temp1
```

```python
df['coverage'] = df['Review'].apply(find_sentence)
```

# obtaining the sentiment score

```python
import vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
df["VADER_Bag_sentiments"] = df['coverage'].apply(analyzer.polarity_scores)

l = []
for i in range(len(df)):
    z = df['VADER_Bag_sentiments'][i].get('compound')
    l.append(z)
    
df['Fa_compound']=l
```

# Obtain all the sentences for the parameter service

```python

def find_sentence(review):
    service=['appointment','years','information','medical','doctors','surgery','service','work','member','experience','family','system','insurance','paid','test','patients','billing','payment']
    temp = []
    for word in service:
        blob = TextBlob(review)
        sentences = blob.sentences
        for sentence in sentences:
            if word in sentence.split():
                temp.append(sentence)
    temp1 = ''
    for i in temp:
        temp1 = temp1+str(i)
    return temp1
```

```python
df['service'] = df['Review'].apply(find_sentence)
```

```python
df['service'] 
```

# obtain the sentiments score for the parameter service

```python
df["Compound_Score_Service"] = df['service'].apply(analyzer.polarity_scores)

l = []
for i in range(len(df)):
    z = df['Compound_Score_Service'][i].get('compound')
    l.append(z)
    
df['Fa_compound_service_score']=l
```
